```python
from WeiboCrawler import *
from utils import *
from Agents import Coordinator, SentimentAnalysistAgent, TopicModellingAgent, Summarizer
import json
import datetime
import os
import pandas as pd
from tqdm import tqdm
import csv
import matplotlib.pyplot as plt
```

```python
def conversation_loop():
    """
    ä¸ç”¨æˆ·æŒç»­å¯¹è¯ï¼Œç›´åˆ° agent è¿”å›åŒ…å«æ‰€æœ‰å¿…éœ€ä¿¡æ¯çš„ JSON æ ¼å¼å›å¤ï¼Œ
    æ­¤æ—¶è°ƒç”¨å¾®åšçˆ¬è™«ï¼Œå¹¶è¿”å›ç”Ÿæˆçš„ CSV æ–‡ä»¶åå’Œäº‹ä»¶å…³é”®å­—ã€‚
    """
    coordinator = Coordinator(prompt_filepath='./prompts/Coordinator_prompt.txt')
    
    while True:
        user_text = input("è¯·è¾“å…¥æŸ¥è¯¢å†…å®¹ï¼š")
        # è°ƒç”¨ Coordinator å¤„ç†ç”¨æˆ·è¾“å…¥
        response = coordinator.run(user_text)
        response = clean_json_output(response)
        print("Agent å›å¤ï¼š", response)
        
        # å°è¯•è§£æå›å¤ä¸º JSON æ ¼å¼
        try:
            result = json.loads(response)
            required_keys = ["event_keywords", "start_year", "start_month", "start_day", "event_release_platform"]
            if all(key in result for key in required_keys):
                platform = result["event_release_platform"].lower()
                if platform not in ["weibo", "å¾®åš"]:
                    print("ç›®å‰æœåŠ¡ä»…æ”¯æŒå¾®åšå¹³å°ï¼Œè¯·é‡æ–°æä¾›ç›¸å…³ä¿¡æ¯ã€‚")
                    continue

                try:
                    year = int(result["start_year"])
                    month = int(result["start_month"])
                    day = int(result["start_day"])
                except Exception as e:
                    print("æ—¶é—´ä¿¡æ¯æ ¼å¼æœ‰è¯¯ï¼Œè¯·æ£€æŸ¥åé‡è¯•ã€‚", e)
                    continue

                # ä¿å­˜äº‹ä»¶å…³é”®å­—ï¼ˆåˆ—è¡¨æ ¼å¼ï¼‰ï¼Œå–ç¬¬ä¸€ä¸ªå…³é”®å­—ä½œä¸ºçˆ¬è™«å…³é”®å­—
                event_keywords = result["event_keywords"]
                keyword = event_keywords[0].strip("#")
                
                # å›å¤æ ¼å¼å®Œæ•´æ—¶è°ƒç”¨çˆ¬è™«ï¼Œå¹¶è¿”å› CSV æ–‡ä»¶åå’Œäº‹ä»¶å…³é”®å­—
                csv_file = run_weibo_crawl(year, month, day, keyword)
                print("å¾®åšçˆ¬è™«å·²å¯åŠ¨ï¼Œçˆ¬å–ä»»åŠ¡å¼€å§‹æ‰§è¡Œã€‚")
                return csv_file, event_keywords
        except json.JSONDecodeError:
            continue


def perform_sentiment_analysis(csv_file: str):
    """
    è¯»å– CSV æ–‡ä»¶ï¼Œåˆ†æ‰¹è°ƒç”¨æƒ…æ„Ÿåˆ†æ agentï¼Œ
    å¹¶ç´¯è®¡ç»Ÿè®¡æƒ…æ„Ÿç»“æœã€‚åŒæ—¶å°†æ¯æ¡å¾®åšæ–‡æœ¬åŠå¯¹åº”çš„æƒ…æ„Ÿå†™å…¥ "public_opinion.csv" æ–‡ä»¶ã€‚
    è¿”å› sentiment_counts å­—å…¸ã€‚
    """
    df = pd.read_csv(csv_file, encoding='utf-8')
    chunk_size = 10
    num_rows = len(df)
    sentiment_counts = {"positive": 0, "neutral": 0, "negative": 0}
    
    # è¾“å‡ºæ–‡ä»¶å
    output_file = "sentiment_analysis_output.csv"
    # æ£€æŸ¥æ–‡ä»¶æ˜¯å¦å­˜åœ¨ï¼Œè‹¥ä¸å­˜åœ¨åˆ™å…ˆåˆ›å»ºå¹¶å†™å…¥è¡¨å¤´
    if not os.path.exists(output_file):
        with open(output_file, "w", newline='', encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["ç¼–å·", "å¾®åšæ­£æ–‡", "sentiment"])
    
    for start in tqdm(range(0, num_rows, chunk_size), desc="Processing chunks"):
        # æ¯ä¸ªæ‰¹æ¬¡é‡æ–°å®ä¾‹åŒ– agentï¼Œé¿å…å¯¹è¯å†å²è¿‡é•¿
        agent = SentimentAnalysistAgent("./prompts/Sentiment_analysist_prompt.txt")
        
        # å–å‡ºå½“å‰å—æ•°æ®ï¼Œå¹¶ä¸ºæ¯æ¡å¾®åšæ·»åŠ ç¼–å·
        chunk = df.iloc[start:start + chunk_size].copy()
        chunk.reset_index(drop=True, inplace=True)
        chunk['ç¼–å·'] = chunk.index + 1
        
        # æ‹¼æ¥æ¯æ¡å¾®åšæ–‡æœ¬ï¼ˆå‡è®¾ CSV ä¸­çš„â€œå¾®åšæ­£æ–‡â€åˆ—åŒ…å«æ–‡æœ¬ï¼‰
        query_lines = chunk.apply(lambda row: f"{row['ç¼–å·']}: {row['å¾®åšæ­£æ–‡']}", axis=1)
        query = "\n".join(query_lines)
        
        response_str = agent.run(query)
        cleaned_response = clean_json_output(response_str)
        print(cleaned_response)
        
        try:
            response_data = json.loads(cleaned_response)
            # æ›´æ–°ç´¯è®¡æƒ…æ„Ÿç»Ÿè®¡
            summary = response_data.get("summary", {})
            sentiment_counts["positive"] += summary.get("positive", 0)
            sentiment_counts["neutral"] += summary.get("neutral", 0)
            sentiment_counts["negative"] += summary.get("negative", 0)
            
            # è·å–æ¯æ¡å¾®åšå¯¹åº”çš„æƒ…æ„Ÿç»“æœ
            analyses = response_data.get("analyses", [])
            # æ£€æŸ¥ analyses æ•°é‡æ˜¯å¦ä¸å½“å‰ chunk æ•°é‡ä¸€è‡´
            if len(analyses) != len(chunk):
                print("è­¦å‘Šï¼šåˆ†æç»“æœæ•°é‡ä¸åŸå§‹æ•°æ®æ•°é‡ä¸åŒ¹é…ï¼")
            
            # å°†æ¯æ¡å¾®åšåŠå¯¹åº”æƒ…æ„Ÿå†™å…¥æ–‡ä»¶
            with open(output_file, "a", newline='', encoding="utf-8") as f:
                writer = csv.writer(f)
                for i, row in chunk.iterrows():
                    if i < len(analyses):
                        sentiment = analyses[i].get("sentiment", "")
                    else:
                        sentiment = ""
                    writer.writerow([row["ç¼–å·"], row["å¾®åšæ­£æ–‡"], sentiment])
        except json.JSONDecodeError as e:
            print("JSONè§£æå¤±è´¥:", e)

    print("ç´¯è®¡æƒ…æ„Ÿç»Ÿè®¡:")
    print("Positive:", sentiment_counts["positive"])
    print("Neutral:", sentiment_counts["neutral"])
    print("Negative:", sentiment_counts["negative"])
    return sentiment_counts, output_file

def perform_topic_analysis(csv_file: str):
    """
    è¯»å– CSV æ–‡ä»¶ï¼Œåˆ†æ‰¹è°ƒç”¨ TopicModellingAgent è¿›è¡Œä¸»é¢˜åˆ†æï¼Œ
    å¹¶æ ¹æ®è¿”å›ç»“æœæ›´æ–°ä¸»é¢˜åˆ—è¡¨å’Œç»Ÿè®¡æ¯ä¸ªä¸»é¢˜çš„è®¨è®ºé‡ï¼ŒåŒæ—¶å°†æ¯æ¡å…¬æ°‘æ„è§åŠå¯¹åº”çš„ä¸»é¢˜å†™å…¥ "topic_modelling_output.csv" æ–‡ä»¶ã€‚
    è¿”å›æ›´æ–°åçš„ä¸»é¢˜åˆ—è¡¨å’Œç»Ÿè®¡å­—å…¸ã€‚
    """
    initial_topics = []
    df = pd.read_csv(csv_file, encoding='utf-8')
    chunk_size = 10
    num_rows = len(df)
    topics = initial_topics.copy()  # åˆå§‹åŒ–ä¸»é¢˜åˆ—è¡¨
    topic_counts = {}  # åˆå§‹åŒ–æ¯ä¸ªä¸»é¢˜çš„è®¨è®ºé‡ç»Ÿè®¡

    output_file = "topic_modelling_output.csv"
    # æ£€æŸ¥æ–‡ä»¶æ˜¯å¦å­˜åœ¨ï¼Œè‹¥ä¸å­˜åœ¨åˆ™åˆ›å»ºå¹¶å†™å…¥è¡¨å¤´
    if not os.path.exists(output_file):
        with open(output_file, "w", newline='', encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["ç¼–å·", "å¾®åšæ­£æ–‡", "topics"])

    for start in tqdm(range(0, num_rows, chunk_size), desc="Processing chunks"):
        # æ¯ä¸ªæ‰¹æ¬¡é‡æ–°å®ä¾‹åŒ– agentï¼Œé¿å…å¯¹è¯å†å²è¿‡é•¿
        agent = TopicModellingAgent("./prompts/Topic_modelling_prompt.txt")
        
        # å–å‡ºå½“å‰å—æ•°æ®ï¼Œå¹¶ä¸ºæ¯æ¡æ•°æ®æ·»åŠ ç¼–å·
        chunk = df.iloc[start:start + chunk_size].copy()
        chunk.reset_index(drop=True, inplace=True)
        chunk['ç¼–å·'] = chunk.index + 1
        
        # æ‹¼æ¥æ¯æ¡å…¬æ°‘æ„è§æ•°æ®ï¼ˆå‡è®¾ CSV ä¸­çš„â€œå¾®åšæ­£æ–‡â€åˆ—åŒ…å«å…¬æ°‘æ„è§ï¼‰
        query_lines = chunk.apply(lambda row: f"{row['ç¼–å·']}: {row['å¾®åšæ­£æ–‡']}", axis=1)
        opinions_text = "\n".join(query_lines)
        
        # æ„é€  queryï¼šåŒ…å«å½“å‰æ‰¹æ¬¡çš„å…¬æ°‘æ„è§å’Œå·²ç¡®å®šçš„ä¸»é¢˜åˆ—è¡¨
        topics_str = ", ".join(topics) if topics else "æ— "
        query = (
            f"è¯·åŸºäºä»¥ä¸‹å…¬æ°‘æ„è§æ•°æ®å’Œå½“å‰ä¸»é¢˜åˆ—è¡¨è¿›è¡Œä¸»é¢˜åˆ†æï¼š\n\n"
            f"å…¬æ°‘æ„è§ï¼š\n{opinions_text}\n\n"
            f"å½“å‰ä¸»é¢˜åˆ—è¡¨ï¼š{topics_str}\n\n"
        )
        
        response_str = agent.run(query)
        cleaned_response = clean_json_output(response_str)
        print(cleaned_response)
        
        try:
            response_data = json.loads(cleaned_response)
            analyses = response_data.get("analyses", [])
            # å°†æ¯æ¡å…¬æ°‘æ„è§åŠå¯¹åº”çš„ä¸»é¢˜å†™å…¥æ–‡ä»¶
            with open(output_file, "a", newline='', encoding="utf-8") as f:
                writer = csv.writer(f)
                if len(analyses) != len(chunk):
                    print("è­¦å‘Šï¼šåˆ†æç»“æœæ•°é‡ä¸åŸå§‹æ•°æ®æ•°é‡ä¸åŒ¹é…ï¼")
                for i, row in chunk.iterrows():
                    if i < len(analyses):
                        topics_list = analyses[i].get("topics", [])
                        topics_output = ", ".join(topic.strip() for topic in topics_list if topic.strip())
                    else:
                        topics_output = ""
                    writer.writerow([row["ç¼–å·"], row["å¾®åšæ­£æ–‡"], topics_output])
            
            # æ›´æ–°ä¸»é¢˜åˆ—è¡¨å’Œè®¨è®ºé‡ç»Ÿè®¡
            for item in analyses:
                topics_list = item.get("topics", [])
                for topic in topics_list:
                    topic = topic.strip()
                    if topic:
                        if topic not in topics:
                            topics.append(topic)
                            print("æ–°å¢ä¸»é¢˜ï¼š", topic)
                        if topic in topic_counts:
                            topic_counts[topic] += 1
                        else:
                            topic_counts[topic] = 1
        except json.JSONDecodeError as e:
            print("JSONè§£æå¤±è´¥:", e)

    print("æ›´æ–°åçš„ä¸»é¢˜åˆ—è¡¨:")
    print(topics)
    print("æ¯ä¸ªä¸»é¢˜çš„è®¨è®ºé‡:")
    print(topic_counts)
    return topic_counts, output_file

def summarize_sentiment_event(event_keywords, sentiment_counts):
    """
    æ ¹æ®äº‹ä»¶å…³é”®å­—å’Œç´¯ç§¯æƒ…æ„Ÿç»Ÿè®¡ç”Ÿæˆæ€»ç»“çš„ promptï¼Œ
    å¹¶è°ƒç”¨ Summarizer agent è¿”å›æ¸…æ´—åçš„äº‹ä»¶æ€»ç»“ã€‚
    
    å‚æ•°:
        event_keywords: äº‹ä»¶å…³é”®å­—åˆ—è¡¨ï¼ˆä¾‹å¦‚ ["#é‡‘ä»·ä¸Šæ¶¨#"]ï¼Œå–ç¬¬ä¸€ä¸ªå…³é”®å­—ï¼‰
        sentiment_counts: ç´¯ç§¯æƒ…æ„Ÿç»Ÿè®¡å­—å…¸ï¼ŒåŒ…å« "positive", "neutral", "negative"
        
    è¿”å›:
        Summarizer ç”Ÿæˆçš„äº‹ä»¶æ€»ç»“æ–‡æœ¬ï¼ˆæ¸…æ´—åçš„ç»“æœï¼‰
    """
    if isinstance(event_keywords, list):
        event_keyword = event_keywords[0]
    else:
        event_keyword = event_keywords

    prompt = (
        f"è¯·å¸®æˆ‘å¯¹ {event_keyword} äº‹ä»¶è¿›è¡Œæ€»ç»“ï¼Œå®ƒçš„ç´¯è®¡æƒ…æ„Ÿç»Ÿè®¡:\n"
        f"Positive: {sentiment_counts['positive']}\n"
        f"Neutral: {sentiment_counts['neutral']}\n"
        f"Negative: {sentiment_counts['negative']}"
    )
    print(prompt)
    summarizer = Summarizer("./prompts/Summarizer_sentiment.txt")
    summarizer_response = summarizer.run(prompt)
    cleaned_summary_response = clean_json_output(summarizer_response)
    
    return cleaned_summary_response

def summarize_topic_event(event_keywords, topic_counts):
    """
    æ ¹æ®äº‹ä»¶å…³é”®å­—å’Œä¸»é¢˜è®¨è®ºé‡ç”Ÿæˆæ€»ç»“çš„ promptï¼Œ
    å¹¶è°ƒç”¨ Summarizer agent è¿”å›æ¸…æ´—åçš„äº‹ä»¶æ€»ç»“ã€‚

    å‚æ•°:
        event_keywords: äº‹ä»¶å…³é”®å­—åˆ—è¡¨ï¼ˆä¾‹å¦‚ ["#æ·±åœ³ç–«æƒ…#"]ï¼Œå–ç¬¬ä¸€ä¸ªå…³é”®å­—ï¼‰
        topic_counts: ç´¯ç§¯ä¸»é¢˜è®¨è®ºé‡ç»Ÿè®¡å­—å…¸ï¼Œä¾‹å¦‚:
                      {'æ·±åœ³å¸‚æ°‘å¯¹ç–«æƒ…çš„ä¸æ»¡æƒ…ç»ª': 7, 'æ·±åœ³ç–«æƒ…é˜²æ§æªæ–½': 12, ...}
        
    è¿”å›:
        Summarizer ç”Ÿæˆçš„äº‹ä»¶æ€»ç»“æ–‡æœ¬ï¼ˆæ¸…æ´—åçš„ç»“æœï¼‰
    """
    # å–å‡ºäº‹ä»¶å…³é”®å­—ï¼ˆè‹¥ä¸ºåˆ—è¡¨åˆ™å–ç¬¬ä¸€ä¸ªï¼‰
    if isinstance(event_keywords, list):
        event_keyword = event_keywords[0]
    else:
        event_keyword = event_keywords

    # æ„é€ ä¸»é¢˜ç»Ÿè®¡éƒ¨åˆ†çš„æ–‡æœ¬ï¼Œæ¯è¡Œæ ¼å¼ä¸º "ä¸»é¢˜: è®¨è®ºé‡"
    topics_str = ""
    for topic, count in topic_counts.items():
        topics_str += f"{topic}: {count}\n"

    prompt = (
        f"è¯·å¸®æˆ‘å¯¹ {event_keyword} äº‹ä»¶è¿›è¡Œæ€»ç»“ï¼Œä»¥ä¸‹æ˜¯å„ä¸»é¢˜çš„è®¨è®ºé‡ç»Ÿè®¡ï¼š\n"
        f"{topics_str}"
    )
    print(prompt)
    
    # ä½¿ç”¨ Summarizer agentï¼ˆç³»ç»Ÿæç¤ºè·¯å¾„ä¸º "./prompts/Summarizer_topic_sentiment.txt"ï¼‰
    summarizer = Summarizer("./prompts/Summarizer_topic.txt")
    summarizer_response = summarizer.run(prompt)
    cleaned_summary_response = clean_json_output(summarizer_response)
    
    return cleaned_summary_response

def summarize_event(event_keyword, sentiment_summary=None, topic_summary=None, merged_analysis=None):
    """
    åˆå¹¶æƒ…æ„Ÿåˆ†ææ€»ç»“ã€ä¸»é¢˜åˆ†ææ€»ç»“ä»¥åŠåˆå¹¶ç»Ÿè®¡ä¿¡æ¯ï¼Œå¹¶ç”Ÿæˆæœ€ç»ˆäº‹ä»¶æ€»ç»“ã€‚

    å‚æ•°:
        event_keyword: äº‹ä»¶å…³é”®å­—ï¼ˆä¾‹å¦‚ "#æ·±åœ³ç–«æƒ…#"ï¼‰
        sentiment_summary: æƒ…æ„Ÿåˆ†æç”Ÿæˆçš„æ€»ç»“æ–‡æœ¬ï¼ˆå­—ç¬¦ä¸²ï¼Œå¯ä¸º Noneï¼‰
        topic_summary: ä¸»é¢˜åˆ†æç”Ÿæˆçš„æ€»ç»“æ–‡æœ¬ï¼ˆå­—ç¬¦ä¸²ï¼Œå¯ä¸º Noneï¼‰
        merged_analysis: åˆå¹¶ç»Ÿè®¡ä¿¡æ¯ï¼Œå¯ä»¥æ˜¯åŒ…å« topics å¯¹åº” sentiment æ•°é‡ç»Ÿè®¡çš„ DataFrame æˆ–å­—ç¬¦ä¸²

    è¿”å›:
        Summarizer agent ç”Ÿæˆçš„æœ€ç»ˆäº‹ä»¶æ€»ç»“æ–‡æœ¬ï¼ˆæ¸…æ´—åçš„ç»“æœï¼‰
    """
    # å°† merged_analysis è½¬æ¢ä¸ºå­—ç¬¦ä¸²ï¼ˆå¦‚æœå®ƒæ˜¯ DataFrameï¼‰
    if merged_analysis is not None:
        if hasattr(merged_analysis, 'to_string'):
            merged_analysis_str = merged_analysis.to_string()
        else:
            merged_analysis_str = str(merged_analysis)
    else:
        merged_analysis_str = "æ— "
    
    sentiment_text = sentiment_summary if sentiment_summary is not None else "æ— "
    topic_text = topic_summary if topic_summary is not None else "æ— "
    
    prompt = (
        f"è¯·å¸®æˆ‘å¯¹ {event_keyword} äº‹ä»¶è¿›è¡Œæ€»ç»“ï¼Œç»¼åˆä»¥ä¸‹ä¸‰éƒ¨åˆ†ä¿¡æ¯ï¼š\n\n"
        f"ã€æƒ…æ„Ÿåˆ†ææ€»ç»“ã€‘:\n{sentiment_text}\n\n"
        f"ã€ä¸»é¢˜åˆ†ææ€»ç»“ã€‘:\n{topic_text}\n\n"
        f"ã€åˆå¹¶ç»Ÿè®¡ä¿¡æ¯ã€‘:\n{merged_analysis_str}\n\n"
        "è¯·ç”Ÿæˆæœ€ç»ˆçš„äº‹ä»¶æ€»ç»“ã€‚"
    )
    print(prompt)
    
    summarizer = Summarizer("./prompts/Summarizer_topic.txt")
    summarizer_response = summarizer.run(prompt)
    cleaned_summary_response = clean_json_output(summarizer_response)
    
    return cleaned_summary_response

if __name__ == "__main__":
    # ç¬¬ä¸€æ­¥ï¼šå¯¹è¯è·å–å‚æ•°ï¼Œå¹¶å¯åŠ¨å¾®åšçˆ¬è™«
    csv_file, event_keywords = conversation_loop()
    
    if os.path.exists(csv_file):
        # ç¬¬äºŒæ­¥ï¼šè¿›è¡Œæƒ…æ„Ÿåˆ†æï¼Œç´¯è®¡æƒ…æ„Ÿç»Ÿè®¡
        sentiment_counts, sentiment_output = perform_sentiment_analysis(csv_file)
        topic_counts, topic_output = perform_topic_analysis(csv_file)
        # ç¬¬ä¸‰æ­¥ï¼šç”Ÿæˆäº‹ä»¶æ€»ç»“å¹¶è°ƒç”¨ Summarizer ç”Ÿæˆæœ€ç»ˆæ€»ç»“
        sentiment_summary = summarize_sentiment_event(event_keywords, sentiment_counts)
        topic_summary = summarize_topic_event(event_keywords, topic_counts)
        print("sentimentæ€»ç»“å›å¤ï¼š", sentiment_summary)
        print("topicæ€»ç»“å›å¤ï¼š", topic_summary)

        merged_analysis = merge_and_analyze(sentiment_output, topic_output)
        summary = summarize_event(event_keywords, sentiment_summary, topic_counts, print(merged_analysis.to_string()))
        print("ç»¼åˆåˆ†æå›å¤ï¼š", summary)

        
    else:
        print("çˆ¬è™«ç”Ÿæˆçš„ CSV æ–‡ä»¶ä¸å­˜åœ¨ï¼Œè¯·æ£€æŸ¥ã€‚")
```

```python
def perform_topic_analysis(csv_file: str, initial_topics: list):
    """
    è¯»å– CSV æ–‡ä»¶ï¼Œåˆ†æ‰¹è°ƒç”¨ TopicModellingAgent è¿›è¡Œä¸»é¢˜åˆ†æï¼Œ
    å¹¶æ ¹æ®è¿”å›ç»“æœæ›´æ–°ä¸»é¢˜åˆ—è¡¨å’Œç»Ÿè®¡æ¯ä¸ªä¸»é¢˜çš„è®¨è®ºé‡ï¼ŒåŒæ—¶å°†æ¯æ¡å…¬æ°‘æ„è§åŠå¯¹åº”çš„ä¸»é¢˜å†™å…¥ "topic_modelling_output.csv" æ–‡ä»¶ã€‚
    è¿”å›æ›´æ–°åçš„ä¸»é¢˜åˆ—è¡¨å’Œç»Ÿè®¡å­—å…¸ã€‚
    """
    df = pd.read_csv(csv_file, encoding='utf-8')
    chunk_size = 10
    num_rows = len(df)
    topics = initial_topics.copy()  # åˆå§‹åŒ–ä¸»é¢˜åˆ—è¡¨
    topic_counts = {}  # åˆå§‹åŒ–æ¯ä¸ªä¸»é¢˜çš„è®¨è®ºé‡ç»Ÿè®¡

    output_file = "topic_modelling_output.csv"
    # æ£€æŸ¥æ–‡ä»¶æ˜¯å¦å­˜åœ¨ï¼Œè‹¥ä¸å­˜åœ¨åˆ™åˆ›å»ºå¹¶å†™å…¥è¡¨å¤´
    if not os.path.exists(output_file):
        with open(output_file, "w", newline='', encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["ç¼–å·", "å¾®åšæ­£æ–‡", "topics"])

    for start in tqdm(range(0, num_rows, chunk_size), desc="Processing chunks"):
        # æ¯ä¸ªæ‰¹æ¬¡é‡æ–°å®ä¾‹åŒ– agentï¼Œé¿å…å¯¹è¯å†å²è¿‡é•¿
        agent = TopicModellingAgent("./prompts/Topic_modelling_prompt.txt")
        
        # å–å‡ºå½“å‰å—æ•°æ®ï¼Œå¹¶ä¸ºæ¯æ¡æ•°æ®æ·»åŠ ç¼–å·
        chunk = df.iloc[start:start + chunk_size].copy()
        chunk.reset_index(drop=True, inplace=True)
        chunk['ç¼–å·'] = chunk.index + 1
        
        # æ‹¼æ¥æ¯æ¡å…¬æ°‘æ„è§æ•°æ®ï¼ˆå‡è®¾ CSV ä¸­çš„â€œå¾®åšæ­£æ–‡â€åˆ—åŒ…å«å…¬æ°‘æ„è§ï¼‰
        query_lines = chunk.apply(lambda row: f"{row['ç¼–å·']}: {row['å¾®åšæ­£æ–‡']}", axis=1)
        opinions_text = "\n".join(query_lines)
        
        # æ„é€  queryï¼šåŒ…å«å½“å‰æ‰¹æ¬¡çš„å…¬æ°‘æ„è§å’Œå·²ç¡®å®šçš„ä¸»é¢˜åˆ—è¡¨
        topics_str = ", ".join(topics) if topics else "æ— "
        query = (
            f"è¯·åŸºäºä»¥ä¸‹å…¬æ°‘æ„è§æ•°æ®å’Œå½“å‰ä¸»é¢˜åˆ—è¡¨è¿›è¡Œä¸»é¢˜åˆ†æï¼š\n\n"
            f"å…¬æ°‘æ„è§ï¼š\n{opinions_text}\n\n"
            f"å½“å‰ä¸»é¢˜åˆ—è¡¨ï¼š{topics_str}\n\n"
        )
        
        response_str = agent.run(query)
        cleaned_response = clean_json_output(response_str)
        print(cleaned_response)
        
        try:
            response_data = json.loads(cleaned_response)
            analyses = response_data.get("analyses", [])
            # å°†æ¯æ¡å…¬æ°‘æ„è§åŠå¯¹åº”çš„ä¸»é¢˜å†™å…¥æ–‡ä»¶
            with open(output_file, "a", newline='', encoding="utf-8") as f:
                writer = csv.writer(f)
                if len(analyses) != len(chunk):
                    print("è­¦å‘Šï¼šåˆ†æç»“æœæ•°é‡ä¸åŸå§‹æ•°æ®æ•°é‡ä¸åŒ¹é…ï¼")
                for i, row in chunk.iterrows():
                    if i < len(analyses):
                        topics_list = analyses[i].get("topics", [])
                        topics_output = ", ".join(topic.strip() for topic in topics_list if topic.strip())
                    else:
                        topics_output = ""
                    writer.writerow([row["ç¼–å·"], row["å¾®åšæ­£æ–‡"], topics_output])
            
            # æ›´æ–°ä¸»é¢˜åˆ—è¡¨å’Œè®¨è®ºé‡ç»Ÿè®¡
            for item in analyses:
                topics_list = item.get("topics", [])
                for topic in topics_list:
                    topic = topic.strip()
                    if topic:
                        if topic not in topics:
                            topics.append(topic)
                            print("æ–°å¢ä¸»é¢˜ï¼š", topic)
                        if topic in topic_counts:
                            topic_counts[topic] += 1
                        else:
                            topic_counts[topic] = 1
        except json.JSONDecodeError as e:
            print("JSONè§£æå¤±è´¥:", e)

    print("æ›´æ–°åçš„ä¸»é¢˜åˆ—è¡¨:")
    print(topics)
    print("æ¯ä¸ªä¸»é¢˜çš„è®¨è®ºé‡:")
    print(topic_counts)
    return topics, topic_counts
```
